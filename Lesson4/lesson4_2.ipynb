```python
age =int(input("請輸入年齡"))
if age >= 18:
    print("符合法定年齡")
    print("可以購買菸酒")

else:
    print("不符合法定年齡")
    print(f"再等{18-age}年才可以購買菸酒")
print("感謝您")
```

```python
#巢狀判斷

age = 26
if age <18: #第1個if
    print("未滿18")

else:
    if age <=25: #第2個if
        print("18~25之間")
    else:
        print("26歲以上")
```

```python
#多項選擇
age = 26
if age <18: #只有1個if，if規則執行完就跳出，N個區塊只執行1個
    print("未滿18")


elif age <=25:
        print("18~25之間")
else:
        print("26歲以上")
```

```python
score = int(input("請輸入總成績"))
attendance = int(input("請輸入出席率"))

# 第一層判斷：檢查出席率是否合格 (大於等於 90)
if attendance >= 90:
    print("出席率：合格")
    
    # 第二層判斷 (巢狀)：檢查分數等級
    if score >= 90:
        print("成績等級：A (優秀)")
    elif score >= 80:
        print("成績等級：B (良好)")
    elif score >= 60:
        print("成績等級：C (及格)")
    else:
        print("成績等級：D (不及格)")
        
else:
    # 出席率不合格
    print("出席率：不合格，不計算成績等級。")

# 輸出結果
# 出席率：合格
# 成績等級：B (良好)
```
