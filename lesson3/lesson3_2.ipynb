```python
a = 10
b = 3
print(a + b)   # 13
print(a / b)   # 3.333...（浮點數除法）
print(a // b)  # 3     （整數除法，取商）
print(a % b)   # 1     （取餘數）
print(a ** b)  # 1000  （次方）
```
